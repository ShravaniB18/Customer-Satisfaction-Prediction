```python
# PHASE 1: imports & load
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import re
from textblob import TextBlob
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve
import shap
import warnings
warnings.filterwarnings('ignore')

# Path to CSV 
DATA_PATH = "customer_support_tickets.csv"

# load
df = pd.read_csv(DATA_PATH)
print("Loaded:", df.shape)
print(df.columns.tolist())
```

```python
# PHASE 2: cleaning & feature engineering

# 1. Copy original (optional)
data = df.copy()

# 2. Standardize column names if needed (optional)
# data.columns = [c.strip() for c in data.columns]

# 3. Drop exact ID columns we don't want to train on
for c in ['Ticket ID', 'Customer Name', 'Customer Email']:
    if c in data.columns:
        data.drop(columns=[c], inplace=True)

# 4. Create Ticket_Text (combine subject, description, resolution)
text_cols = ['Ticket Subject', 'Ticket Description', 'Resolution']
for c in text_cols:
    if c not in data.columns:
        data[c] = ""

data['Ticket_Text'] = (data['Ticket Subject'].fillna('') + " " +
                       data['Ticket Description'].fillna('') + " " +
                       data['Resolution'].fillna(''))

# 5. Clean text helper (basic)
def clean_text(s):
    if pd.isna(s): return ""
    s = str(s).lower()
    s = re.sub(r"http\S+|www\S+", " ", s)
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

data['Ticket_Text'] = data['Ticket_Text'].map(clean_text)

# 6. Parse dates and compute resolution duration (hours) if times exist
def parse_time_features(df_in):
    out = df_in.copy()
    # Date of Purchase -> year/month
    if 'Date of Purchase' in out.columns:
        out['Date of Purchase'] = pd.to_datetime(out['Date of Purchase'], errors='coerce')
        out['purchase_year'] = out['Date of Purchase'].dt.year.fillna(-1).astype(int)
        out['purchase_month'] = out['Date of Purchase'].dt.month.fillna(-1).astype(int)
    else:
        out['purchase_year'] = -1
        out['purchase_month'] = -1

    # First Response Time and Time to Resolution -> try to parse as datetime or HH:MM
    def to_datetime_or_time(x):
        try:
            return pd.to_datetime(x)
        except:
            # try if format like 'HH:MM' -> attach arbitrary date
            try:
                return pd.to_datetime("2000-01-01 " + str(x))
            except:
                return pd.NaT

    if 'First Response Time' in out.columns:
        fr = out['First Response Time'].apply(to_datetime_or_time)
    else:
        fr = pd.Series(pd.NaT, index=out.index)

    if 'Time to Resolution' in out.columns:
        tr = out['Time to Resolution'].apply(to_datetime_or_time)
    else:
        tr = pd.Series(pd.NaT, index=out.index)

    out['first_response_ts'] = fr
    out['time_to_resolution_ts'] = tr

    dur = (tr - fr).dt.total_seconds() / 3600.0
    out['resolution_duration_hours'] = dur.fillna(-1).replace([np.inf, -np.inf], -1)
    # response hour/day
    out['first_response_hour'] = fr.dt.hour.fillna(-1).astype(int)
    out['first_response_dow'] = fr.dt.dayofweek.fillna(-1).astype(int)
    return out

data = parse_time_features(data)

# 7. Sentiment polarity (TextBlob) â€” simple numeric signal
data['sentiment_polarity'] = data['Ticket_Text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# 8. Numeric cleanup: Customer Age (make numeric)
if 'Customer Age' in data.columns:
    data['Customer Age'] = pd.to_numeric(data['Customer Age'], errors='coerce').fillna(data['Customer Age'].median())

# 9. Target: binary satisfied vs not
SAT_THRESHOLD = 3  # change to 3 if you prefer >=3 as satisfied
target_col = 'Customer Satisfaction Rating'
# drop rows with missing target
data = data.dropna(subset=[target_col])
# convert to numeric if needed
data[target_col] = pd.to_numeric(data[target_col], errors='coerce')
data.dropna(subset=[target_col], inplace=True)
data[target_col] = data[target_col].astype(float)
data['Satisfied'] = (data[target_col] >= SAT_THRESHOLD).astype(int)

# 10. Keep only useful columns (important features)
# We pick: Ticket_Text (NLP), Ticket Priority, Ticket Type, Ticket Channel, Customer Age, resolution_duration_hours, sentiment_polarity
keep_cols = [
    'Ticket_Text',
    'Ticket Priority', 'Ticket Type', 'Ticket Channel',
    'Customer Age', 'resolution_duration_hours', 'sentiment_polarity',
    'purchase_year', 'purchase_month', 'Satisfied'
]
# Ensure columns exist
keep_cols = [c for c in keep_cols if c in data.columns]
data = data[keep_cols].copy()

print("After FE shape:", data.shape)
data.head()
```

```python
# ================= PHASE 5: Model Training & Evaluation =================
from xgboost import XGBClassifier, DMatrix, train as xgb_train
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import joblib
import xgboost as xgb

# Convert to arrays 
X_train_part, X_val, y_train_part, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print("â³ Transforming data...")
X_train_t = preprocessor.fit_transform(X_train_part)
X_val_t = preprocessor.transform(X_val)
X_test_t = preprocessor.transform(X_test)
print("Transformation complete!")

# Convert to DMatrix (XGBoost's optimized data format)
dtrain = xgb.DMatrix(X_train_t, label=y_train_part)
dval = xgb.DMatrix(X_val_t, label=y_val)
dtest = xgb.DMatrix(X_test_t, label=y_test)

# Define parameters manually (works for all versions)
params = {
    'objective': 'multi:softprob',
    'num_class': len(np.unique(y_train)),
    'eval_metric': 'mlogloss',
    'learning_rate': 0.05,
    'max_depth': 8,
    'subsample': 0.9,
    'colsample_bytree': 0.9,
    'min_child_weight': 3,
    'reg_lambda': 2,
    'nthread': -1,
    'random_state': 42,
    'verbosity': 1
}

# Train using xgb.train() (supports early stopping in all versions)
evals = [(dtrain, 'train'), (dval, 'val')]
print("Training with early stopping...")
bst = xgb_train(
    params=params,
    dtrain=dtrain,
    num_boost_round=1000,
    evals=evals,
    early_stopping_rounds=30,
    verbose_eval=50
)
print("Training complete!")

# Make predictions
y_pred_probs = bst.predict(dtest)
y_pred = np.argmax(y_pred_probs, axis=1)

# Evaluate results
acc = accuracy_score(y_test, y_pred)
print(f"\n Test Accuracy: {acc:.4f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Save model & preprocessor
joblib.dump(bst, "best_customer_satisfaction_model.pkl")
joblib.dump(preprocessor, "preprocessor.pkl")
print("\n Model & preprocessor saved successfully!")
```
